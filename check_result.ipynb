```python
import json
from IPython.display import HTML
import html

data = 'part1'
model = 'lunar-deepseek-r1'
random_trial = 0
skip_planner = False

selected_sample_ind = 2

if skip_planner:
    result_file = f'./results/results_{data}_skip_planner_{model}_0.json'
else:
    result_file = f'./results/results_{data}_{model}_{random_trial}.json'

with open(result_file, 'r') as f:
    results = json.load(f)

result = results[selected_sample_ind]

display(HTML(f'<b style="font-size:16px;">Problem Information</b><br><br>'
             f'<b>Problem ID:</b> {result.get("problem_id")}<br>'
             f'<b>Problem Type:</b> {result.get("gt_problem_type")}<br>'
             f'<b>Premise:</b> {result.get("premise")}<br>'
             f'<b>Statement:</b> {result.get("statement")}<br>'
             f'<b>Groundtruth Label:</b> {result.get("label")}<br>'
             ))

try:
    execution_str = html.escape(result.get("excution_dag"))
except:
    execution_str = result.get("excution_dag")
display(HTML(f'<b style="font-size:16px;">Planning output</b><br><br>'
            f'<b>Predicted Problem Type:</b> {result.get("pred_problem_type")}<br>'
            f'<b>Reason:</b> {str(result.get("reason"))}<br>'
            f'<b>Selected Solver:</b> {str(result.get("selected_agents"))}<br>'
            f'<b>Plan Execution:</b> {execution_str}<br>'
            ))

solution_final = result.get("solution_final", {})
if len(solution_final) > 0:
    display(HTML(f'<b style="font-size:16px;">Solver Output</b><br><br>'
             f'<b>Prediction Label:</b> {solution_final.get("pred")}<br>'
             f'<b>Is Label Correct:</b> {str(result.get("is_correct"))}<br>'
             f'<b>Reasoning:</b> {str(solution_final.get("reasoning"))}<br>'
             f'<b>Fact Verification:</b> {str(solution_final.get("fact_verification"))}<br>'
             f'<b>Reason Pattern Verification:</b> {str(solution_final.get("reasoning_verification"))}<br>'
             ))
else:
    display(HTML(f'<b style="font-size:16px;">No solution generated.</b><br><br>'))


solution_history = result.get("solution_history", {})
if len(solution_history) > 1:
    solution_history = solution_history[0]
    display(HTML(f'<b style="font-size:16px;">Solution Before Refinement</b><br><br>'
                f'<b>Prediction Label:</b> {solution_history.get("pred")}<br>'
                f'<b>Reasoning:</b> {str(solution_history.get("reasoning"))}<br>'
                f'<b>Fact Verification:</b> {str(solution_history.get("fact_verification"))}<br>'
                f'<b>Reason Pattern Verification:</b> {str(solution_history.get("reasoning_verification"))}<br>'
                ))
else:
    display(HTML(f'<b style="font-size:16px;">No refinement performed.</b><br><br>'))

error_info = result.get("error_info", {})   
if len(error_info) > 1:
    formatted_json = json.dumps(error_info, indent=4)
    display(HTML(f'<b style="font-size:16px;">Error occurred in this sample</b><br><br>'
                 f'<pre style="padding:10px;">{html.escape(formatted_json)}</pre>'
                 ))
else:
    display(HTML(f'<b style="font-size:16px;">No error occurred.</b><br><br>'))

```
